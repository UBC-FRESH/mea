```python
startyear = 2015
fig, ax = plt.subplots(figsize=(6, 4))
df1 = mea_results['net_revenue'].loc['CBM_Base_HWP_Base', :, startyear:]['sum_tnr']
df2 = mea_results['net_revenue'].loc['CBM_A_HWP_Base', :, startyear:]['sum_tnr']

df1.groupby('year').sum().plot(ax=ax, linestyle=':', alpha=0.5)
df2.groupby('year').sum().plot(ax=ax, linestyle='--', alpha=0.5)

ax.legend(['Base scenario', 'Alternate scenario'])
ax.set_xlabel('Time (year)')
ax.set_ylabel('Net revenue ($)')
plt.savefig('biosafe.pdf', format='pdf')
```
